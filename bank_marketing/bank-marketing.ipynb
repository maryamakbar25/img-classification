```python
bank_df["married"] = np.where(bank_df["marital"]=="married", 1, 0)
bank_df["admin"] = np.where(bank_df["job"]=="admin.", 1, 0)
bank_df["edu"] = np.where(bank_df["education"]=="university.degree", 1, 0)
bank_df["defaulted"] = np.where(bank_df["default"]=="yes", 1, 0)
bank_df["housing_loan"] = np.where(bank_df["housing"]=="yes", 1, 0)
bank_df["personal_loan"] = np.where(bank_df["loan"]=="yes", 1, 0)
bank_df["cellular"] = np.where(bank_df["contact"]=="cellular", 1, 0)
bank_df["y_new"] = np.where(bank_df["y"]=="yes", 1, 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split

y = bank_df['y_new'].to_numpy()
x = bank_df.drop(columns=['y','job','education', 'marital','contact','day_of_week','month','poutcome','default','housing','loan']).to_numpy()

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.1, random_state=42)

tree = DecisionTreeClassifier()
fit = tree.fit(x_train, y_train)


fit.get_n_leaves
```

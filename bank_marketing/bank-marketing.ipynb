## Introduction



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

## Dataset



```python
# Load in the dataset
path = Path.cwd() / "bank-additional-full.csv"

bank_df = pd.read_csv(path, sep = ";", header = [0], skipinitialspace= True, index_col = None)
```

```python
bank_df.head()
```

```python
font = { 'family' : 'serif', 
            'color' : 'black'
}
```

```python
fig, ax = plt.subplots(2, sharex= True)

fig.suptitle("Age")

ax[0].hist(bank_df["age"], bins = 20, color="cadetblue")

ax[1].boxplot(bank_df["age"], vert= False)
```

```python
plt.hist(bank_df["job"], bins = 20, color="cadetblue")
plt.title("Most call recipients are either admin or blue collar workers", fontdict=font)
plt.xlabel("Job" , fontdict=font)
plt.xticks(rotation = 70)
```
